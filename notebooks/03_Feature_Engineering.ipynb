# Feature Engineering

## Objective

Create meaningful business features that improve demand prediction and pricing analysis.

The engineered features will be used in the machine learning model and price elasticity analysis.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/data/retail_price.csv")
```

## Revenue

Revenue represents the total income generated from product sales.

Formula:

Revenue = Quantity Sold × Unit Price

```python
df["revenue"] = df["qty"] * df["unit_price"]

df[["qty","unit_price","revenue"]].head()
```

## Average Competitor Price

Average competitor price provides an overall market price for comparison.

```python
df["avg_comp_price"] = (
    df["comp_1"] +
    df["comp_2"] +
    df["comp_3"]
) / 3

df[["comp_1","comp_2","comp_3","avg_comp_price"]].head()
```

## Price Difference

Price Difference measures how much higher or lower our price is compared to competitors.

```python
df["price_difference"] = (
    df["unit_price"] -
    df["avg_comp_price"]
)

df[["unit_price","avg_comp_price","price_difference"]].head()
```

```python
df["price_ratio"] = (
    df["unit_price"] /
    df["avg_comp_price"]
)

df[["unit_price","avg_comp_price","price_ratio"]].head()
```

```python
df["freight_percentage"] = (
    df["freight_price"] /
    df["unit_price"]
) * 100

df[["freight_price","unit_price","freight_percentage"]].head()
```

```python
df["revenue_per_customer"] = (
    df["revenue"] /
    df["customers"]
)

df[["revenue","customers","revenue_per_customer"]].head()
```
